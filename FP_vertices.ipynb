```python
# Importing libraries
import numpy as np
import struct

def float_to_ieee_754_16bit(num):
    # Convert the float number to 16-bit half-precision float using NumPy
    np_half_float = np.float16(num)
    #print(np_half_float)

    # Pack the 16-bit half-precision float as binary data (2 bytes)
    packed = struct.pack('e', np_half_float)
    #print(packed)

    # Unpack it as a 16-bit unsigned integer
    ieee_754_16bit = struct.unpack('H', packed)[0]
    #print(ieee_754_16bit)

    # Return the binary representation as a string
    binary_format = format(ieee_754_16bit, '016b')
    #Grouping the binary string as 1 bit, then 5 bits, then 10 bits
    group_1 = binary_format[:1]    # First 1 bit
    group_5 = binary_format[1:6]   # Next 5 bits
    group_10 = binary_format[6:16] # Last 10 bits

#Joining the groups with spaces in between
    return f'{group_1} {group_5} {group_10}'

def matrix_to_ieee_754_16bit(matrix):
    # Convert each element in the matrix
    ieee_matrix = np.vectorize(float_to_ieee_754_16bit)(matrix)
    return ieee_matrix
```

```python
# Step 1: Open and read file
with open('teapot.data', 'r') as file:
    data = file.readlines()
```

```python
# Step 2: Parse the contents and store the numbers in a list
numbers_list = []
for line in data:
    numbers = line.strip().split()
    numbers_list.extend([float(num) for num in numbers])
```

```python
# Step 3: Converting the list to a NumPy array and reshape it into an 8x3 matrix

matrix = np.array(numbers_list).reshape(306, 3)
print(matrix)
```

```python
# Printing the IEEE representation matrix
ieee_754_matrix = matrix_to_ieee_754_16bit(matrix)
'''print(type(ieee_754_matrix[0][1]))
print(ieee_754_matrix[0][0]*2)'''
print(ieee_754_matrix)
```

```python
#Storing values in a .txt file
with open('matrix_input.txt', 'w') as f:
    # Write matrix A
    for row in ieee_754_matrix:
        f.write(' '.join(map(str, row)) + '\n')
```
